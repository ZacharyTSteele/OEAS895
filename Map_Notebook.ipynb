```python
import cartopy as cp
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

```python
# create a blank global map

fig = plt.figure(figsize = (10,8), dpi = 300)

ax = fig.add_subplot(1,1,1, projection = ccrs.InterruptedGoodeHomolosine())
ax.set_global()

ax.coastlines()

ax.add_feature(cp.feature.OCEAN, facecolor = 'pink', alpha= 0.75)
ax.add_feature(cp.feature.LAND, facecolor = 'black')

fig.savefig('Work_Of_Art.png', dpi=300)
```
