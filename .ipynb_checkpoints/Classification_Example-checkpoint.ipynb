```python
%matplotlib inline

import numpy as np
from sklearn.datasets import load_iris

import matplotlib.pyplot as plt
```

```python
iris = load_iris()

iris
```

```python
iris.feature_names
```

```python
iris.target
```

```python
iris.data
```

```python
fig =plt.figure(1, figsize =(4,3), dpi = 300)

plt.scatter(iris.data[:,0], iris.data[:,2], s=50, c=iris.target, label=iris.target_names)
plt.ylabel(iris.feature_names[2])
plt.xlabel(iris.feature_names[0])
plt.colorbar()


fig.show
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

from sklearn import metrics
```

```python
x = iris.data
y = iris.target

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4,random_state=4) #test size set to 20% and selected random state 4
```

```python
len(x_train)
```

```python
len(y_train)
```

```python
y_train.shape
```

```python
x_train.shape
```

```python
knn = KNeighborsClassifier(n_neighbors =5)

knn.fit(x_train,y_train)

y_pred = knn.predict(x_test)

print('accuracy score =', metrics.accuracy_score(y_test, y_pred))
```

```python
metrics.confusion_matrix(y_test, y_pred)
```

```python
print(metrics.classification_report(y_test, y_pred))
```
